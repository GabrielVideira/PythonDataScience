```python
tmdb = pd.read_csv('tmdb_5000_movies.csv')
tmdb.head(10)
```

# <font color='red' style='font-size: 30px;'>Análises Preliminares</font>
<hr style='border: 2px solid red;'>

```python
tmdb.describe().round(2)
```

```python
ax = sns.distplot(tmdb.vote_average)
ax.set(xlabel='Nota Média', ylabel='Densidade')
ax.set_title('Media de votos em filmes no TMDB 5000')
```

```python
ax = sns.distplot(tmdb.vote_average, norm_hist= False, kde = False)
ax.set(xlabel='Nota Média', ylabel='Densidade')
ax.set_title('Media de votos em filmes no TMDB 5000')
```

```python
ax = sns.boxplot(x=tmdb.vote_average)
```

#### <font color='red'>Observação</font>
***
> É possível ver nesse gráfico uma densidade anormal de ocorrências para Nota **Média de 0**.

> Analisando o dataset vi que a quantidade anormal ocorrências com Nota Média 0 deve-se **por terem filmes sem nenhum voto, logo sua nota média segue 0 ou por terem menos de 5 votos.**

```python
tmdb.query('vote_average == 0')
```

```python
tmdb_com_mais_de_10_votos = tmdb.query('vote_count >= 10')
tmdb_com_mais_de_10_votos.describe()
```

```python
ax = sns.distplot(tmdb_com_mais_de_10_votos.vote_average)
ax.set(xlabel='Nota Média', ylabel='Densidade')
ax.set_title('Media de votos em filmes no TMDB 5000')
```

```python
ax = sns.distplot(tmdb_com_mais_de_10_votos.vote_average, kde_kws={'cumulative':True}, hist_kws={'cumulative': True})
ax.set(xlabel='Nota Média', ylabel='% Acumulada de Filmes')
ax.set_title('Media de votos em filmes no TMDB 5000')
```
