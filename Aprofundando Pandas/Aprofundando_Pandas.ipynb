## **Separar os tipos de residencia em sua variável. Exemplo: Tipo Casa fica armazenado numa variavel casa.**

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python

'''

Funçoes lambda sao funçoes anonimas onde podemos declarar uma funçao sem dar
nome a ela
Elas são úteis quando precisamos de uma função simples, que será utilizada
apenas uma vez e não precisa ser reutilizada em outro lugar do código

lambda {argumentos}: {expressão}
'''
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
dados
```

# **Indentificando e Removendo Outliers**

```python

import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (14,6))
```

```python
dados[dados['Valor'] >= 500000]
```

```python
valor = dados.Valor
```

```python
Q1 = valor.quantile(.25)
Q3 = valor.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 *IIQ
```

```python
selecao = (valor >=limite_inferior)&(valor<= limite_superior)
dados_new = dados[selecao]
```

## **Continuando Outliers**

```python
dados.boxplot(['Valor'], by = ['Tipo'])
```

```python
grupo_tipo = dados.groupby('Tipo')['Valor']
```

```python
type(grupo_tipo)
```

```python
grupo_tipo.groups
```

```python
Q1 = grupo_tipo.quantile(.25)
Q3 = grupo_tipo.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 *IIQ
```

```python
Q1
```

```python
for tipo in grupo_tipo.groups.keys():
    print(tipo)
```

```python
for tipo in grupo_tipo.groups.keys():
    eh_tipo = dados['Tipo'] == tipo
    eh_dentro_limite = (dados['Valor'] >= limite_inferior[tipo]) & (dados['Valor'] <= limite_superior[tipo])
    selecao = eh_tipo & eh_dentro_limite
    dados[selecao]
```

```python
dados_new = pd.DataFrame()
for tipo in grupo_tipo.groups.keys():
    eh_tipo = dados['Tipo'] == tipo
    eh_dentro_limite = (dados['Valor'] >= limite_inferior[tipo]) & (dados['Valor'] <= limite_superior[tipo])
    selecao = eh_tipo & eh_dentro_limite
    dados_selecao = dados[selecao]
    dados_new = pd.concat([dados_new, dados_selecao])
```

```python
dados_new.boxplot(['Valor'], by = ['Tipo'])
```

```python
dados_new.to_csv('aluguel_residencial_sem_outliers.csv', sep = ';', index = False)
```
