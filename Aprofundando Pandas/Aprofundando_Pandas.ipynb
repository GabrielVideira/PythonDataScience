```python
import pandas as pd
dados = pd.read_csv('aluguel.csv', sep =';')
```

```python
dados_residencial = dados[selecao] # Aqui excluimos os valores q estao armazenadas como False em 'selecao'
```

```python
dados_residencial
```

```python
list(dados_residencial['Tipo'].drop_duplicates()) # Verificando se realmente filtrou por locações residenciais
```

# **Relatório de Análise IV**

Mostre a frequencia dos imóveis classificados com tipo 'Apartamento'.

Mostre a frequencia dos imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'.

Mostre a frequencia dos imóveis com área entre 60 e 100 metros quadrados, incluindo os limites.

Mostre a frequencia dos imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00.

```python
import pandas as pd

dados = pd.read_csv('aluguel.csv', sep = ';')
```

```python
dados.head(10)
```

```python
selecao = dados['Tipo'] == 'Apartamento'
selecao
```

```python
# Mostre a frequencia dos imóveis classificados com tipo 'Apartamento'.

selecao = dados['Tipo'] == 'Apartamento'
n1 = dados[selecao].shape[0]
n1
```

```python
# Mostre a frequencia dos imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'.

selecao = (dados['Tipo'] =='Casa')|(dados['Tipo'] == 'Casa de Condomínio')|(dados['Tipo']=='Casa de Vila')
n2 = dados[selecao].shape[0]
n2
```

```python
# Mostre a frequencia dos imóveis com área entre 60 e 100 metros quadrados, incluindo os limites.

selecao = (dados['Area'] >= 60) & (dados['Area'] <= 100 )
n3 = dados[selecao].shape[0]
n3
```

```python
# Mostre a frequencia dos imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00.

selecao = (dados['Quartos'] >= 4) & (dados['Valor']< 2000)
n4 = dados[selecao].shape[0]
n4
```

# **Relatório de Análise V**

## **Tratamento de dados faltantes**

```python
import pandas as pd
dados = pd.read_csv('aluguel_residencial.csv', sep =';')
```

```python
dados.head(10)
```

```python
dados.isnull()
```

```python
dados.notnull()
```

```python
dados.info() # Conseguimos verificar ali quantos valores sao nao nulos por indice, assim sabemos tb oq tratar
```

```python
dados['Valor'].isnull()
```

```python
dados[dados['Valor'].isnull()]
```

```python
# dropna, serve pra remover devido subeset com as condiçoes impostas
A = dados.shape[0]
dados.dropna(subset = ['Valor'], inplace = True)
B = dados.shape[0]
A - B
```

```python
dados[dados['Valor'].isnull()]
```

```python
''' Tratando a coluna condominio, seguindo a regra de negocio em que todos as residencias
    do tipo apartamento devem conter um valor de condominio'''
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
```

```python
A = dados.shape[0]
dados = dados[~selecao] # nessa linha, passamos a serie
B = dados.shape[0]
A - B
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
dados.fillna(0, inplace = True) # Para preencher valores NaN/Nulos
```

```python
dados.info()
```

```python
dados.to_csv('aluguel_residencial.csv', sep =';', index = False)
```
