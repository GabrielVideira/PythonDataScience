# **Relatório de Análise III**

## **Imóveis residenciais**

```python
import pandas as pd
dados = pd.read_csv('aluguel.csv', sep =';')
```

```python
list(dados['Tipo'].drop_duplicates())
```

```python
residencial = ['Quitinete',
'Casa',
'Apartamento',
'Casa de Condomínio',
'Casa de Vila'] # Definindo conjunto de locaçoes residenciais
```

```python
residencial
```

```python
selecao = dados['Tipo'].isin(residencial) # Funçao isin retorna True ou False se tiver aquele dado salvo.
selecao.head(10)
```

```python
dados_residencial = dados[selecao] # Aqui excluimos os valores q nao estao armazenadas no DataFrame 'selecao'
```

```python
dados_residencial
```

```python
list(dados_residencial['Tipo'].drop_duplicates()) # Verificando se realmente filtrou por locações residenciais
```

## **Exportando a Base de Dados**

```python
dados_residencial.to_csv('aluguel_residencial.csv', sep=';')
```
